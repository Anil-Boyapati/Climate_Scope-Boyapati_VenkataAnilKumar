```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include='object').columns
```

```python
#univariate analysis

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
# For Categorical Columns
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Frequency of {col}')
    plt.show()
```

```python
#bivariate

numeric_cols = df.select_dtypes(include=['number']).columns

# Avoid empty dataframes
if len(numeric_cols) > 1:
    plt.figure(figsize=(12,8))
    sns.heatmap(df[numeric_cols].corr(), cmap='coolwarm')
    plt.title("Correlation Heatmap")
    plt.show()
else:
    print("⚠️ Not enough numeric columns for correlation heatmap.")
```

```python
#detecting the outliers
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#removing duplicates

df = df.drop_duplicates()
print("Duplicates removed. New shape:", df.shape)
```

```python
#handling missing data

df.fillna(df.mean(numeric_only=True), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
print("Missing values handled.")
```

```python
# data types conversion

df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')

# Example: Convert sunrise/sunset from string to time
for col in ['sunrise', 'sunset']:
    df[col] = pd.to_datetime(df[col], format='%I:%M %p', errors='coerce').dt.time

print("Data type conversion done.")
```

```python
# standadizing units

df.drop(['temperature_fahrenheit', 'wind_mph', 'pressure_in', 'precip_in',
         'visibility_miles', 'gust_mph'], axis=1, inplace=True, errors='ignore')

print("Units standardized and redundant columns removed.")
```

```python
#features name changing

df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['day'] = df['last_updated'].dt.day
df['hour'] = df['last_updated'].dt.hour

# Example: Categorize temperature
df['temp_category'] = pd.cut(df['temperature_celsius'],
                             bins=[-50, 0, 15, 25, 35, 50],
                             labels=['Freezing', 'Cold', 'Mild', 'Warm', 'Hot'])
print("Feature engineering completed.")
```

```python
#filtering noisy data

df = df[df['temperature_celsius'].between(-50, 60)]
df = df[df['humidity'].between(0, 100)]

print("Filtered noisy/unrealistic data.")
```

```python
#aggregrating data

monthly_avg = df.groupby(['country', 'year', 'month'])['temperature_celsius'].mean().reset_index()
print("Aggregated monthly averages created.")
display(monthly_avg.head())
```

```python
#saving the cleaned data

df.to_csv("weather_cleaned.csv", index=False)
print("✅ Final cleaned dataset saved as 'weather_cleaned.csv'")
```
